```python
%matplotlib inline
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.3, random_state=0)
poly = PolynomialFeatures(degree=3)
lr = LogisticRegression()

# plot the original data
plt.figure()
plt.plot(X[np.where(y==1)[0], 0], X[np.where(y==1)[0], 1], 'o')
plt.plot(X[np.where(y==0)[0], 0], X[np.where(y==0)[0], 1], 's', c = 'r')

# build  a LR classifier
lr.fit(X, y)

# generate some data to test the model; remember to transform the features
# then classify the new data 
Xe, ye = make_moons(n_samples=3000, noise=0.3, random_state=0)
yhat_lr = lr.predict(Xe)

# plot the decision boundary of our classifier
plt.figure()
plt.plot(Xe[np.where(yhat_lr==1)[0], 0], Xe[np.where(yhat_lr==1)[0], 1], 'o')
plt.plot(Xe[np.where(yhat_lr==0)[0], 0], Xe[np.where(yhat_lr==0)[0], 1], 's', c = 'r')
```
